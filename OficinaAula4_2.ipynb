Importando Bibliotecas

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

Lendo base de dados

```
iris = sns.load_dataset("mpg")
iris.head()
```

### **Visualização que mostre a distribuição conjunta de todas as variáveis quantitativas, diferenciando pelo país de origem (‘origin’)**

```
sns.pairplot(iris, hue='origin')
```

### **Quais variáveis possuem uma maior correlação?**

Nota-se que todas as variavéis possuem maior correlação com a variável 'model_year' e menor correlação com a variável cylinders.

### **Sobre a visualização da questão 2, é possível distinguir o país de origem a partir da distribuição de uma variável? E de uma distribuição conjunta?**

Observa-se, nos gráficos a respeito da distribuição conjunta, que os países de origem são diferenciados pela cor, como informando na legenda. A função pairplot() é permite, através de uma única função, conseguimos uma visão geral de todas as variáveis do conjunto de dados. Para essa diferenciação, nesse caso, basta passar a coluna ‘origin’ para o parâmetro hue. 

### **Escolha uma distribuição conjunta em que as variáveis estejam fortemente correlacionadas e crie um gráfico que permita analisar a distribuição conjunta, a reta de regressão e as distribuições marginais.**

```
sns.jointplot(x="mpg", y="model_year", kind = 'reg', data=iris)
```

**Gráfico extra

```
sns.pairplot(iris, hue='origin', kind = 'reg')
```

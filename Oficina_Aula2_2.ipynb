## **Oficina referente à Aula 2.2 do Curso de Visualização de Dados**

Importando Bibliotecas

```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Lendo Base de Dados

```
df = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt')
df
```

## **Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo;**

Criando meu dataset apenas com os dados da empresa Nintendo

```
graf1 = df[df['Publisher']=='Nintendo']
graf1
```

Agrupando os dados em torno dos gêneros e calculando o somatório do total de vendas em que cada gênero 

```
generos = graf1.groupby(by='Genre').sum()
generos
```

Plotando o gráfico

```
plt.figure(figsize=(10,10))
plt.style.use('seaborn-pastel')
plt.pie(x=generos['Global_Sales'], labels = generos.index, autopct='%1.1f%%')
plt.show()
```

## **Uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015**

Criando meu dataset com informações a partir do ano de 2015

```
graf2 = df[df['Year']>=2015.0]
graf2
```

Verificando a quantidade de vezes que a quantidade de jogos aparece por Platform

```
platform = graf2.groupby(by='Platform').count()
```

Plotando o Gráfico

```
plt.figure(figsize=(10,10))
plt.pie(x=platform['Year'], labels = platform.index, autopct='%1.1f%%')
plt.show()
```

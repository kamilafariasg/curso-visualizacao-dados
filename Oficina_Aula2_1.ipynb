Importando bibliotecas necessárias 

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Lendo base de dados

```
filmes_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

## **Questão 2: Visualização que mostre o orçamento de todos os filmes do conjunto de dados**

Ordenando conjunto de dados pelo orçamento:'Budget_M'

```
filmes = filmes.sort_values(by=['Budget_M'], axis=0)
```

Plotando o gráfico que irá representar o orçamento de todos os filmes.

```
sns.set(style="darkgrid")
plt.figure(figsize=(20,30))
plt.title('Orçamento de todos os filmes')
sns.barplot(y="Film", x='Budget_M', color = 'r', estimator = sum, data=filmes)
plt.show()
```

**Qual o filme com maior orçamento e com menor orçamento?**

Através do gráfico, é possível notar que o filme com menor orçamento é o primeiro: "The Twilight Saga: New Moon". E o filme com maior orçamento é o último: "Pirates of the Caribbean: At World's End"

## **Questão 3: Visualização que mostre a bilheteria de todos os filmes do conjunto de dados**

Ordenando conjunto de dados pelo orçamento:'Worldwide_Gross_M'

```
filmes = filmes.sort_values(by=['Worldwide_Gross_M'], axis=0)
```

Plotando o gráfico que irá representara bilheteria de todos os filmes.

```
sns.set(style="darkgrid")
plt.figure(figsize=(20,30))
plt.title('Orçamento de todos os filmes')
sns.barplot(y="Film", x='Worldwide_Gross_M', color = 'orange', estimator = sum, data=filmes)
plt.show()
```

**Qual o filme com maior bilheteria e com menor bilheteria?**

O gráfico simboliza a quantidade de bilheteria de cada filme ordenado em ordem crescente. A visualização de dados permite identicar claramente que o filme "Avatar" tem a maior quantidade de bilheteria. E o filme "Ratatouille" com o menor número de bilheteria.

## **Questão 4: Uma visualização que mostre a quantidade de filmes por gênero em cada ano.**

```
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', data=filmes)
plt.show()
```

**Em que anos foram feitos menos filmes de ação?**

Atraves da legenda, podemos identificar que os filmes de ação são representados pela cor azul. Logo, o ano que representa essa menor quantidade, são os anos de 2008 e 2010;

 **E em que ano foram feitos menos filmes de animação?**

Os filmes de animação, por sua vez, são identificados pela cor verde, como podemos verificar na legenda do gráfico. Assim, o ano que representa a menor quantidade é o ano de 2011;

## **Questão 5: Uma visualização que mostre a média de orçamento de filmes por gênero em cada ano.**

```
plt.figure(figsize=(15,8))
plt.title('Média de orçamento por gênero em cada ano')
sns.barplot(x="Year", y = 'Budget_M' , ci = None, hue= 'Genre', data=filmes)
plt.show()
```

**Em que ano os filmes de ação tiveram maior média de orçamento?**

Através da legenda, podemos identificar que os filmes de ação são representados pela cor azul. Logo, o ano que representa essa maior quantidade de orçamento, é o ano de 2007;

**E em que ano os filmes de animação tiveram a maior média de orçamento?**

Os filmes de animação, por sua vez, são identificados pela cor verde, como podemos verificar na legenda do gráfico. Assim, o ano que representa a maior quantidade de orçamento é o ano de 2009;

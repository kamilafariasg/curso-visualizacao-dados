Importando Bibliotecas

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

Lendo Base de Dados

```
uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt'
df = pd.read_csv(uri)
df.head()
```

### **Crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade.**

```
plt.figure(figsize=(20,5))
plt.title('Variação de vendas ao longo do tempo através de meses')
sns.lineplot(x="Month", y="Sales", data=df)
plt.show()
```

A tendência de uma série temporal é definida como um padrão de crescimento/descrecimento da variável em um certo período de tempo.Através do gráfico de séries temporais podemos notar que existe uma tendência positiva. Já a periodicidade, são padrões durante o comportamento da série. Analisando o gráfico notamos que o número de venda tende a subir durante o final do ano, ou seja, existe tbm uma periodicidade.

Importando bibliotecas

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

Lendo Base de Dados

```
df = sns.load_dataset('mpg')
df.head()
```

### **Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem?**

```
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo do combustível pelo país de origem')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=df)
plt.show()
```

Nota-se que  existe uma tendência do valor da aceleração crescer, quando o valor do consumo cresce. Observamos baixa correlação positiva para os Estados Unidos. Sem correlação para a Europa e correlação positiva para o Japão.

### **Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?**

```
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo do combustível pelo ano do modelo')
sns.scatterplot(x="acceleration", y="mpg", hue = 'model_year', data=df)
plt.show()
```

Por padrão de cores do seaborn, os valores mais baixos são mapeados em cores mais claras, e os mais altos em cores mais escuras. Assim, notamos uma certa correlação positiva entre a aceleração, o valor do consumo e ano do modelo.

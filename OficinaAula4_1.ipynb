Importando Bibliotecas

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

Lendo Base de Dados

```
titanic = sns.load_dataset("titanic")
titanic.head()
```

### **Visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro embarcou ("embark_town").**

```
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.stripplot(x="embark_town", y="fare", data=titanic)
plt.show()
```

Através do gráfico, podemos notar que a cidade de Queenstown tem o número mais inferior em relação ao valor da passagem. Já a cidade de Cherboung apresenta valores maiores. 

### **Visualização que mostre as distribuições do valor da passagem, de acordo com a cidade de embarque, sexo e se o passageiro sobreviveu ou não**

```
sns.catplot(x="fare", y="embark_town", hue = 'sex', col = 'alive',kind = 'swarm', data=titanic)
```

**Que diferença você consegue notar nas distribuições dos passageiros que sobreviveram em relação aos que não sobreviveram?**

Nota-se que a maioria dos sobreviventes são do sexo feminino. Além disso, observa-se, também, que existem mais passageiros na cidade de Southampton. Outro fato, é que existem mais passageiros onde o preço da passagem é menor que 100.
